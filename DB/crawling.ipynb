```python

import requests
from bs4 import BeautifulSoup
import re

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
cnt = 0
for i in range(1,2):
    response = requests.get(f"https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={i}",  headers=headers)  
    #header를 안넣으면 봇으로 알고 접속을 끊어버린다.
    #params에도 '?'뒤의 페이지 정보를 넣을 수 있음
    soup = BeautifulSoup(response.text, 'html.parser')
    brand = soup.select('#searchList .li_inner .article_info > .item_title')
    title = soup.select('#searchList .li_inner .article_info > .list_info')
    price = soup.select('#searchList .li_inner .article_info > .price')
    for i in range(len(brand)):
        try:
            _price = price[i].text.strip().split('\n')
            a = {"브랜드": brand[i].text.strip(), "제목":title[i].text.strip(), "정상가": _price[0], "할인가":_price[-1].strip()}
            print(a)
            cnt += 1
        except:
            pass
cnt

    
    
```

```python
import requests
from bs4 import BeautifulSoup
import re

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
dicts = {}
for i in range(1,2):
    response = requests.get(f"https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=437&listType=title&date=20230712&page={i}",  headers=headers)  #header를 안넣으면 봇으로 알고 접속을 끊어버린다. 
    soup = BeautifulSoup(response.text, 'html.parser')
    print(len(soup.select('.type02 a')))
    for i in range(len(soup.select('.type02 a'))):
        title = soup.select('.type02 a')[i].text
        data = soup.select('.type02 a')[i].attrs['href'].split(' ')
        response_2 = requests.get(data[0], headers=headers)
        soup_2 = BeautifulSoup(response_2.text, 'html.parser')
        _data = soup_2.select('#newsct_article')
        dicts[title] = _data
print(dicts)
 

```
