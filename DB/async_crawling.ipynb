### 비동기 방식 크롤링

```python
import requests
from bs4 import BeautifulSoup
url ="http://www.cine21.com/rank/boxoffice/domestic"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
movies = soup.select('#boxoffice_list')
movies
```

```python
url = 'http://www.cine21.com/rank/boxoffice/domestic_content'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
data = {
'page': 1,
'genre': 'all',
'period': '2023-07-07'
}  # post방식에는 이렇게 데이터를 담아야한다. get은 선택사항
response = requests.post(url, data=data, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
li_tags = soup.select('.boxoffice_li')
for li_tag in li_tags:
    print(li_tag.select_one('.mov_name').text.strip())
```

```python
url = "https://comic.naver.com/api/webtoon/titlelist/weekday?order=user"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}

response = requests.get(url ,  headers = headers)
data = response.json()
sunday = data['titleListMap']['SUNDAY']
monday = data['titleListMap']['MONDAY']
tuesday = data['titleListMap']['TUESDAY']
wednesday = data['titleListMap']['WEDNESDAY']
thursday = data['titleListMap']['THURSDAY']
friday = data['titleListMap']['FRIDAY']
saturday = data['titleListMap']['SATURDAY']
week = [sunday, monday, wednesday, tuesday, thursday, friday, saturday]
# for key in data['titleListMap'] 이렇게 하면 영어로 요일을 하나씩 쓸필요가 없ㅂ다. 
dict = {}
cnt = 0
for i in week:
    for j in range(len(i)):  # 범위로 하기보다는 data['titleListMap'][key]에서 하나씩 꺼낸다.
        dict[f'제목{cnt}'] = i[j]['titleName']
        dict[f'썸네일{cnt}'] = i[j]['thumbnailUrl']
        cnt += 1  #리스트 안에 딕셔너리를 넣으면 굳이 이름을 안바궈도 된다.

dict


# for key in data['titleListMap']:
#     for webtoon in data['titleListMap'][key]:
#         webtoon['titleName']
#         webtoon['thumbnailUrl']


```
