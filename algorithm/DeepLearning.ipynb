```python
from sklearn.datasets import load_wine
import numpy as np
import pandas as pd

data_wine = load_wine()

wine_df = pd.DataFrame(data_wine.data, columns =data_wine.feature_names)
target_df = pd.DataFrame(data_wine.target, columns = ['target'])

def euclidean_distance(a, b):
    dist = np.sqrt(sum((a-b)**2))
    return dist

def cosine_similarity(a, b):
    return np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))

data = []
data2 = []
for i in range(len(wine_df)):
    _data = []
    _data2 = []
    # 기준 데이터
    for j in range(len(wine_df)):   # 비교 대상
        # distance 계산
        # similarity 계산
        base_point = wine_df.iloc[i].values
        target_point = wine_df.iloc[j].values
        _data.append(euclidean_distance(base_point, target_point))
        _data2.append(cosine_similarity(base_point, target_point))
    data.append(_data)
    data2.append(_data2)

pd.DataFrame(data, columns=list(range(178)))    
#pd.DataFrame(data2, columns=list(range(178)))    
```
